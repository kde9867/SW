```python
import pandas as pd
import jinja2 
from pycaret.regression import *
from pycaret.classification import *
from time import time
import os

data_path = '/content/sample_data'
```

```python
train = pd.read_csv(os.path.join(data_path, '/content/sample_datafinal_train (5).csv'))
test = pd.read_csv(os.path.join(data_path, '/content/sample_datafinal_test (4).csv'))

display(train.head())
display(test.head())
```

```python
cell_start_time = time()
clf = setup(data = train, target='nerdiness', silent=True)
cell_end_time = time()
print("CELL RUN TIME : ",cell_end_time - cell_start_time)
```

```python
cell_start_time = time()
top5_models = compare_models(fold = 5, round = 5, sort = 'AUC', n_select = 5)
cell_end_time = time()
print("CELL RUN TIME : ",cell_end_time - cell_start_time)
```

```python
top5_models
```

```python
total_models = []
models()
```

```python
cell_start_time = time()
model_catboost = create_model('catboost', fold = 5)
total_models.append(model_catboost)
cell_end_time = time()
print("CELL RUN TIME : ",cell_end_time - cell_start_time)
model_catboost
```

```python
cell_start_time = time()
model_xgboost = create_model('xgboost', fold = 5)
total_models.append(model_xgboost)
cell_end_time = time()
print("CELL RUN TIME : ",cell_end_time - cell_start_time)
model_xgboost
```

```python
cell_start_time = time()
model_rf = create_model('rf', fold = 5)
total_models.append(model_rf)
cell_end_time = time()
print("CELL RUN TIME : ",cell_end_time - cell_start_time)
model_rf
```

```python
cell_start_time = time()
model_et = create_model('et', fold = 5)
total_models.append(model_et)
cell_end_time = time()
print("CELL RUN TIME : ",cell_end_time - cell_start_time)
model_et
```

```python
cell_start_time = time()
model_lightgbm = create_model('lightgbm', fold = 5)
total_models.append(model_lightgbm)
cell_end_time = time()
print("CELL RUN TIME : ",cell_end_time - cell_start_time)
model_lightgbm
```

```python
model_catboost = tune_model(model_catboost, fold=5, optimize = 'AUC', choose_better = True)
```

```python
model_xgboost = tune_model(model_xgboost, fold=5, optimize = 'AUC', choose_better = True)
```

```python
model_rf = tune_model(model_rf, fold=5, optimize = 'AUC', choose_better = True)
```

```python
model_et = tune_model(model_et, fold=5, optimize = 'AUC', choose_better = True)
```

```python
model_lightgbm = tune_model(model_lightgbm, fold=5, optimize = 'AUC', choose_better = True)
```

```python
tuned_models = [model_catboost, model_xgboost, model_rf, model_et, model_lightgbm]
```

```python
for model in tuned_models:
    display(predict_model(model))
```
